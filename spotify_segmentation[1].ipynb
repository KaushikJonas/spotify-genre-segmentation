```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()
label_encoders = {}

# Only encode columns of type 'object' (strings)
for col in df_encoded.select_dtypes(include=['object']).columns:
    # Check if all values are strings (skip columns with mixed types)
    if df_encoded[col].apply(lambda x: isinstance(x, str)).all():
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col])
        label_encoders[col] = le

df_encoded.head()
```

```python
# Select only numeric columns for correlation
numeric_df = df_encoded.select_dtypes(include=[np.number])
```

```python
# Plot a heatmap of the correlation matrix for numeric_df
plt.figure(figsize=(14, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
#Implementing KMeans clustering
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(numeric_df[['track_popularity', 'duration_ms']])
```

```python
numeric_df['cluster'] = kmeans.labels_
```

```python
numeric_df
```

```python
plt.scatter(numeric_df['track_popularity'], numeric_df['duration_ms'], c=numeric_df['cluster'], cmap='viridis', alpha=0.6)
plt.xlabel('Track Popularity')
plt.ylabel('Duration (ms)')
plt.title('KMeans Clustering of Tracks')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Recommending similar tracks using Nearest Neighbors
import warnings
from sklearn.neighbors import NearestNeighbors

# We'll use the numeric_df for recommendations (excluding the 'cluster' column)
features = numeric_df.drop(columns=['cluster'])

# Fit NearestNeighbors model
nn_model = NearestNeighbors(n_neighbors=6, metric='euclidean')
nn_model.fit(features)

def recommend_tracks(track_index, n_recommendations=5):
    distances, indices = nn_model.kneighbors([features.iloc[track_index]], n_neighbors=n_recommendations+1)
    # Exclude the first result (itself)
    recommended_indices = indices[0][1:]
    return df.iloc[recommended_indices][['track_name', 'track_artist', 'track_album_name', 'track_popularity']]
warnings.filterwarnings("ignore")
# Example: Recommend tracks similar to the first track
recommend_tracks(0)
```
